```python
import pandas as pd
import io

#データを読み込む
input_data = pd.read_csv("Airbnb_Open_Data.csv")
```

```python
input_data.head(10)
```

```python
input_data.info()
```

```python
input_data.describe()
```

```python
input_data['price'] = input_data['price'].str.replace('$', '')
input_data['price'] = input_data['price'].str.replace(',', '')
print(input_data.price)
```

```python
input_data['service fee'] = input_data['service fee'].str.replace('$', '')
input_data['service fee'] = input_data['service fee'].str.replace(',', '')
print(input_data["service fee"])
```

```python
input_data.info()
```

```python
input_data['price'] = pd.to_numeric(input_data['price'], errors="coerce")
input_data['service fee'] = pd.to_numeric(input_data['service fee'], errors="coerce")
# input_data['price'] = input_data['price'].astype('int')
input_data.info()
```

```python
input_data.describe()
```

```python
#価格の平均をもとめる
print(input_data.price.mean())
#価格のメディアン値をもとめる
print(input_data.price.median())
#価格の分散をもとめる
print(input_data.price.var())
#価格の標準偏差をもとめる
print(input_data.price.std())
#価格の最大値をもとめる
print(input_data.price.max())
#価格の最小値をもとめる
print(input_data.price.min())
```

```python
print(input_data["room type"])
```

```python
#ルームタイプごとに、価格の平均をもとめる
print(input_data.groupby('room type')['price'].mean())
#ルームタイプごとに、価格のメディアン値をもとめる
print(input_data.groupby('room type')['price'].median())
#ルームタイプごとに、価格の分散をもとめる
print(input_data.groupby('room type')['price'].var())
#ルームタイプごとに、価格の標準偏差をもとめる
print(input_data.groupby('room type')['price'].std())
#ルームタイプごとに、価格の最大値をもとめる
print(input_data.groupby('room type')['price'].max())
#ルームタイプごとに、価格の最小値をもとめる
print(input_data.groupby('room type')['price'].min())
```

```python
#価格の平均をもとめる
print((input_data["price"]+input_data["service fee"]).mean())
#価格のメディアン値をもとめる
print((input_data["price"]+input_data["service fee"]).median())
#価格の分散をもとめる
print((input_data["price"]+input_data["service fee"]).var())
#価格の標準偏差をもとめる
print((input_data["price"]+input_data["service fee"]).std())
#価格の最大値をもとめる
print((input_data["price"]+input_data["service fee"]).max())
#価格の最小値をもとめる
print((input_data["price"]+input_data["service fee"]).min())
```

```python
import matplotlib.pyplot as plt

#価格(price) についてヒストグラムを描く
plt.hist(input_data.price)
plt.xlabel('price')
plt.ylabel('count')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import scipy.stats

plt.scatter(input_data["minimum nights"],input_data.price, s=50, c="blue", alpha=0.3, linewidths=1, edgecolors="blue")
plt.title('minimum nights and Price')
plt.ylabel('Price')
plt.xlabel('Minimum Nights')
```

```python
import seaborn as sns
sns.jointplot(input_data["minimum nights"],input_data.price)
```

```python
import numpy as np

#モデルで利用する変数を接合し, ひとつの DataFrame にする
#読み込むデータを変えれば, 説明変数や被説明変数を変えることが可能
x_list = pd.concat([input_data.price, input_data["Construction year"],input_data["number of reviews"], input_data["calculated host listings count"], input_data["reviews per month"]], axis=1)

#pairplotをプロットする
sns.pairplot(x_list, kind='reg')
```
